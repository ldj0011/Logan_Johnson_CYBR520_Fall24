```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
data = pd.read_csv("spambase(2).csv")

# Set default figure size
plt.rcParams['figure.figsize'] = (14, 10)

# Select only numeric columns
numeric_data = data.select_dtypes(include='number')
```

```python
# 1. Create a line chart to visualize the distribution of the 'capitalLong' feature.

# Plot each point with markers and connect them with lines
plt.plot(data.index, data['capitalLong'])

# Add labels and title
plt.title('Line Chart of the dispersion of capitalLong')
plt.xlabel('Index')
plt.ylabel('capitalLong')

# Display the graph
plt.show()
```

```python
# 2. Generate a bar chart to show the frequency of "type"(spam and nonspam) in the dataset

# Plot the Data
plt.figure()
sns.countplot(x='type', data=data, hue='type', palette={'spam': 'red', 'nonspam': 'blue'}, legend=False)

# Add Labels and Title
plt.title('Frequency of Spam and Non-Spam')
plt.xlabel('Type')
plt.ylabel('Frequency')

#Display the Graph
plt.show()
```

```python
# 3. Create a scatter plot to explore the relationship between 'num3d' and 'num000'.

# Plot the Data
plt.figure()
sns.scatterplot(x='num3d', y='num000', data=data)

# Add Labels and Title
plt.title('Relationship between num3d and num000')
plt.xlabel('num3d')
plt.ylabel('num000')

#Display the Graph
plt.show()
```

```python
# 4. Visualize the distribution of 'money' using a histogram.

#Plot the Data
plt.figure()
sns.histplot(data['money'], kde=False)

# Adding labels and title
plt.title('Distribution of Money')
plt.xlabel('Money')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# 5. Create a pie chart to represent the proportion of spam and nonspam emails in the dataset.

# Plotting the pie chart
plt.figure()
data['type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['blue','red'])

# Adding title
plt.title('Proportion of Spam and Non-Spam Emails')
plt.ylabel('')  # To hide the y-axis label

# Display the plot
plt.show()
```

```python
#16. Generate a correlation matrix for all numeric features in the dataset

# Create the correlation matrix
correlation_matrix = numeric_data.corr()

# Print the Results
print("Correlation Matrix:")
print(correlation_matrix)

```

```python
#17. Identify the two features with the highest positive correlation and the two with the highest negative correlation.

# Get the absolute values of the correlation matrix
abs_corr_matrix = correlation_matrix.abs()

# Mask the diagonal (self-correlation) by setting it to zero
np.fill_diagonal(abs_corr_matrix.values, 0)

# Find the two features with the highest positive correlation
highest_positive_corr = abs_corr_matrix.unstack().sort_values(ascending=False)
top_positive = highest_positive_corr.head(2)

# Find the two features with the highest negative correlation
highest_negative_corr = correlation_matrix.unstack().sort_values()
top_negative = highest_negative_corr.head(2)

# Display the results
print("Two features with the highest positive correlation:")
print(top_positive)

print("\nTwo features with the highest negative correlation:")
print(top_negative)
```

```python
#18. Visualize the correlation matrix using a heatmap

# Draw the heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,         # Show correlation values
    fmt=".2f",         # Format the numbers to two decimal places
    cmap='coolwarm',   # Color map
    square=True,       # Make cells square-shaped
    cbar_kws={"shrink": .8},  # Shrink color bar
    linewidths=0.5,   # Lines between cells
    linecolor='white'  # Color of the lines
)

# Set the title
plt.title('Correlation Matrix of Numeric Features', fontsize=16)

# Show the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
#24. Explain when it is appropriate to use a scatter plot for data exploration

#Scatter plots are appropriate to use when we want to see trends in data or see the association between data that is continious.
#They are also useful for visualizing outliers that are significantly different from other data points.
```

```python
#25. Discuss the advantages and disadvantages of using a pie chart for categorical data

#Advantages: Pie charts are visually easy to understand how data contributes to the "whole" of a pie. They allow us to quickly compare
#different catergories of data by looking at the "slice" of each category in the pie chart, which makes them more appealing when presenting 
#data to an audience.

#Disadvantages: Pie charts are not best when wanting to display a large amount of catergories and work best when there are 7 or fewer
#categories. They can make it harder to compare categories especially if there are small differences which may lead to misinterpretation of data.
```

```python
#26. When would you choose a line chart over a histogram for data visualization?

#Line charts are best used over a histogram when wanting to display trends of continious variables over a period time, such as the daily 
#temperature or fluctuation of currency exchange rates.
```
