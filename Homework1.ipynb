```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
data = pd.read_csv("Dataset for HW1.csv")

# Set default figure size
plt.rcParams['figure.figsize'] = (14, 10)
```

```python
# 1. Create a line chart to visualize the distribution of the 'capitalLong' feature.

# Plot each point with markers and connect them with lines
plt.plot(data.index, data['capitalLong'])

# Add labels and title
plt.title('Line Chart of the dispersion of capitalLong')
plt.xlabel('Index')
plt.ylabel('capitalLong')

# Display the graph
plt.show()
```

```python
# 2. Generate a bar chart to show the frequency of "type"(spam and nonspam) in the dataset

# Plot the Data
plt.figure()
sns.countplot(x='type', data=data, hue='type', palette={'spam': 'red', 'nonspam': 'blue'}, legend=False)

# Add Labels and Title
plt.title('Frequency of Spam and Non-Spam')
plt.xlabel('Type')
plt.ylabel('Frequency')

#Display the Graph
plt.show()
```

```python
# 3. Create a scatter plot to explore the relationship between 'num3d' and 'num000'.

# Plot the Data
plt.figure()
sns.scatterplot(x='num3d', y='num000', data=data)

# Add Labels and Title
plt.title('Relationship between num3d and num000')
plt.xlabel('num3d')
plt.ylabel('num000')

#Display the Graph
plt.show()
```

```python
# 4. Visualize the distribution of 'money' using a histogram.

#Plot the Data
plt.figure()
sns.histplot(data['money'], kde=False)

# Adding labels and title
plt.title('Distribution of Money')
plt.xlabel('Money')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# 5. Create a pie chart to represent the proportion of spam and nonspam emails in the dataset.

# Plotting the pie chart
plt.figure()
data['type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['blue','red'])

# Adding title
plt.title('Proportion of Spam and Non-Spam Emails')
plt.ylabel('')  # To hide the y-axis label

# Display the plot
plt.show()
```

```python
# 6. Generate a table chart to display the summary statistics for the 'capitalAve' feature.

capitalAve_summary = data['capitalAve'].describe()
print("Summary Statistics for capitalAve:")
print(capitalAve_summary.to_frame().T)
```

```python
# 7. Calculate the mean, median, and standard deviation of the 'our' feature

# Calculate mean, median, and standard deviation
mean_value = data['our'].mean()
median_value = data['our'].median()
std_dev_value = data['our'].std()

# Print the results
print(f"Mean of 'our': {mean_value}")
print(f"Median of 'our': {median_value}")
print(f"Standard Deviation of 'our': {std_dev_value}")
```

```python
# 8. Determine the minimum and maximum values of 'hp'

# Calculate minimum and maximum values
min_value = data['hp'].min()
max_value = data['hp'].max()

# Print the results
print(f"Minimum value of 'hp': {min_value}")
print(f"Maximum value of 'hp': {max_value}")
```

```python
# 9. Calculate the 25th, 50th, and 75th percentiles for 'telnet'

# Calculate the 25th, 50th, and 75th percentiles
percentiles_telnet = np.percentile(data['telnet'], [25, 50, 75])

# Print the results
print("\n25th percentile of 'telnet':", percentiles_telnet[0])
print("50th percentile (median) of 'telnet':", percentiles_telnet[1])
print("75th percentile of 'telnet':", percentiles_telnet[2])
```

```python
# 10.  Find the mode of the 'you' feature.

# Calculate the mode
mode_you = data['you'].mode()

# Print the results
print("\nMode of 'you':", mode_you)
```

```python
# 11. Calculate the interquartile range (IQR) for 'business'.

# Calculate the interquartile range (IQR)
Q1 = data['business'].quantile(0.25)
Q3 = data['business'].quantile(0.75)
IQR = Q3 - Q1

# Print the Results
print("\nInterquartile Range (IQR) for 'business':", IQR)
```

```python
# 12. Compute the skewness and kurtosis of the 'credit' feature.

# Compute the skewness and kurtosis
skewness_credit = data['credit'].skew()
kurtosis_credit = data['credit'].kurtosis()

# Print the Results
print("\nSkewness of 'credit':", skewness_credit)
print("Kurtosis of 'credit':", kurtosis_credit)
```

```python
# 13. Calculate the correlation coefficient between 'capitalAve' and 'capitalTotal'.

# Calculate the correlation coefficent
correlation_capital = data['capitalAve'].corr(data['capitalTotal'])

# Print the Results
print("\nCorrelation coefficient between 'capitalAve' and 'capitalTotal':", correlation_capital)
```

```python
# 14. Determine the range of values for 'num857'.

# Range of Values
range_num857 = data['num857'].max() - data['num857'].min()

# Print the Results
print("\nRange of values for 'num857':", range_num857)
```

```python
# 15. Calculate the variance of 'money'.

# Calculate the Variance
variance_money = data['money'].var()

# Print the Results
print("\nVariance of 'money':", variance_money)
```
